```python
from thesaurusrex import mcPair
import pandas as pd
```

```python
starters = [
'Headache',
 'Blood.',
 'hope',
 'ego',
 'water  ',
 'dream',
 'cat',
 'cloud',
 'paparazzi',
 'dream',
 'wrestlemania',
 'salad',
 'forgiveness',
 'pencil',
 'echo',
 'sheet music',
 'mirror',
 'inkpen',
 'book',
 'lock',
 'volcano',
 'Promise',
 'parrot',
 'yarn',
 'Match.',
 'tomorrow',
 'sieve',
 'bin.',
 'air',
 'mind',
 'battery',
 'wig',
 'weighty',
 'toothpick',
 'mouse',
 'ice cream',
 'snowman',
 'news',
 'spider',
 'x-ray',
 'well',
 'clone',
 'shemale',
 'owl',
 'tomorrow',
 'toenail',
 'February',
 'Olympic',
 'infinity',
 'clock']
```

```python
wa_riddles = []
wa_answers = []
```

```python
for starter in starters:
    searchTerm = starter
    
    try:
        ep = mcPair(searchTerm,2)
        pairs = ep.pairs
        riddle = f"What is as {pairs[0][0]} as {pairs[0][1]} and as {pairs[1][0]} as {pairs[1][1]}?"  
    except:
        riddle = ""
    wa_riddles.append(riddle)
    wa_answers.append(searchTerm)
    
```

```python
for ans, riddle in zip(wa_answers, wa_riddles):
    if len(riddle)!= 0:
        print(ans,'|', riddle)
```
